```python
browser = webdriver.Chrome('C:\\Users\\HP\\AppData\\Roaming\\Python\\Python39\\site-packages\\chromedriver_binary\\chromedriver.exe')

browser.get('https://www.youtube.com/@BBKiVines/videos')  # whenever this command run link will be load
```

```python
# set the scrolling limit
scroll_limit = 100000

# Scroll down the page by 1000 pixels each time
while scroll_limit > 0:
    browser.execute_script("window.scrollBy(0,1000)")
    scroll_limit -= 1000
    
# wait for page to load
time.sleep(5)

soup = BeautifulSoup(browser.page_source,'html.parser')
```

```python
Data = []
for sp in soup.find_all('ytd-rich-grid-row'):
    for s in sp.find_all('ytd-rich-item-renderer'):
        Title = s.find('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get('title')
        video_link = "https://www.youtube.com"+s.find('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get('href')
        views = s.find_all('span',class_="inline-metadata-item style-scope ytd-video-meta-block")[0].text[:-6]
        publish = s.find_all('span',class_="inline-metadata-item style-scope ytd-video-meta-block")[1].text
        Duration = s.find('span',class_ = 'style-scope ytd-thumbnail-overlay-time-status-renderer').text.replace('\n','').strip()
        Data.append([Title,publish,views,video_link,Duration])

print(len(Data))

    
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
Data2 =[]
for x in Data:
    browser.get(x[3])
    # # set the scrolling limit
    # scroll_limit = 200

    # # Scroll down the page by 1000 pixels each time
    # while scroll_limit > 0:
    #     browser.execute_script("window.scrollBy(0,200)")
    #     scroll_limit -= 200
    
    # # wait for page to load
    time.sleep(3)

    
    btn = browser.find_element(By.ID,"expand")
    btn.click()

    # wait for the page to fully load
    time.sleep(3)

    scroll_limit = 8000
    # Scroll down the page by 1000 pixels each time
    while scroll_limit > 0:
        browser.execute_script("window.scrollBy(0,1000)")
        scroll_limit -= 1000
    
    # wait for page to load
    time.sleep(3)

    soup = BeautifulSoup(browser.page_source,'html.parser')
    views = soup.find_all('span',class_ = 'style-scope yt-formatted-string bold')[0].text[:-6]
    Publish_date = soup.find_all('span',class_ = 'style-scope yt-formatted-string bold')[2].text
    try:
        comments = soup.find('yt-formatted-string',class_ = 'count-text style-scope ytd-comments-header-renderer').text[:-9]
    except:
        pass
    try:
        likes = soup.find_all('span','yt-core-attributed-string yt-core-attributed-string--white-space-no-wrap')[3].text
    except:
        pass
    if(likes == 'Subscribe'):
        likes = soup.find_all('span','yt-core-attributed-string yt-core-attributed-string--white-space-no-wrap')[4].text
    Data2.append([views,Publish_date,comments,likes])
```
